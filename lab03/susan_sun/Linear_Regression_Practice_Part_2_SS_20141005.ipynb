```python
print data['lstat'].shape
X = data['lstat'].reshape(506,1)
print X.shape
```

```python
print data['medv'].shape
Y = data['medv'].reshape(506,1)
print Y.shape
```

```python
mod_lstat = LinearRegression(fit_intercept = True)
mod_lstat.fit(X, Y)
X_test = np.linspace(0,40,num=40).reshape(40,1)
y_hat = mod_lstat.predict(X_test)
plt.plot(X, Y, 'o')
plt.plot(X_test, y_hat)
```
