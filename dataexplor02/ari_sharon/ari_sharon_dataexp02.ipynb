```python
x = africa['Infrared02']
m = x.as_matrix()
x1 = m.reshape(x.size,1)
x1[0:5]
```

```python
y = africa['pH']
m2 = y.as_matrix()
y1 = m2.reshape(y.size,1)
y1[0:5]
```

```python
plt.plot(x1,y1,'o')
```

```python
model = LinearRegression(fit_intercept=True)
model.fit(x1, y1)
print ("Model coefficient: %.5f, and intercept: %.5f"
       % (model.coef_, model.intercept_))
```

```python
l =  model.predict(x1)
```
